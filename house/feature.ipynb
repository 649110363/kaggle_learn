#####  以上内容对特征有了比较直观的理解，现在对数据进行预处理，首先是缺失值的处理

```python
total = df_train.isnull().sum().sort_values(ascending=False) #统计缺失值
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)
```

```python
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

#####  缺失值达到90%以上的可能是并不重要的特征，应予以删除，如PoolQC、MiscFeature、Alley等（通常缺失超过15%时应删除相应变量）
##### GarageX系列具有同样的缺失值，由于表示车库信息的字段是GarageCars,不存在缺失值，所以我们可以将该系列字段删除
##### 同样的还有Bsmt系列
##### 而MasVnrType和MasVnrArea则好像并没有多大贡献
##### Electrical 只有一个缺失值，可以删除记录或者取众数，在这里还是删除记录比较好

```python
#缺失值处理
df_train = df_train.drop((missing_data[missing_data['Total'] > 1]).index, axis=1) #去掉total>1的列
df_train = df_train.drop(df_train.loc[df_train['Electrical'].isnull()].index)
```

##### 现在缺失值已经处理完毕，需要注意离群点也要处理

```python
#通过‘SalePrice’的标准偏差和散点图进行分析
#建立一个阈值，将观测定义为异常值，所以要先标准化数据，即将数据值转换为平均值为0，标准偏差为1
from sklearn.preprocessing import StandardScaler
saleprice_scaled = StandardScaler().fit_transform(df_train['SalePrice'][:, np.newaxis])
low_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][:10] #排序时要加[:,0]，将所有元素列出来排序
high_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][-10:]
```

```python
print('--------low range:\n', low_range)
print('--------high range:\n', high_range) # 低范围值离0相似且不太远，高范围值最大为7.多，离0很远
```

##### 以上是单变量分析，下面进行双变量分析

```python
#双变量分析，分别进行saleprice与OverallQual，GrLivArea，TotalBsmtSF等进行分析
var = 'OverallQual'
data = pd.concat([df_train['SalePrice'], df_train[var]] ,axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0, 800000)) #可以看出大致呈线性关系或者二次曲线关系,下面寻找正态性
```

```python
# 直方图和正态概率图
from scipy.stats import norm
from scipy import stats
sns.distplot(df_train['SalePrice'], fit=norm)
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)
```

```python
df_train['SalePrice'] = np.log(df_train['SalePrice']) #将数据转换成log形式
```

```python
sns.distplot(df_train['SalePrice'], fit=norm)
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)
```

```python
#同理分析OverallQual
sns.distplot(df_train['OverallQual'], fit=norm)
fig = plt.figure()
res = stats.probplot(df_train['OverallQual'], plot=plt)
```

```python
sns.distplot(df_train['GrLivArea'], fit=norm)
fig = plt.figure()
res = stats.probplot(df_train['GrLivArea'],plot=plt)
```

```python
df_train['GrLivArea'] = np.log(df_train['GrLivArea'])
```

```python
sns.distplot(df_train['GrLivArea'], fit=norm)
fig = plt.figure()
res = stats.probplot(df_train['GrLivArea'],plot=plt)
```

```python
sns.distplot(df_train['TotalBsmtSF'], fit=norm)
fig = plt.figure()
res = stats.probplot(df_train['TotalBsmtSF'],plot=plt)
```

```python
#由于存在大量0值，故要创建一个变量获得有或没有地下室（二进制变量）的效果，然后将所有非零观察值进行对数转换，忽略零值，这样既可以转换数据又不会失去有或没有地下室的影响
df_train['HasBsmt'] = pd.Series(len(df_train['TotalBsmtSF']), index=df_train.index)
df_train['HasBsmt'] = 0
df_train.loc[df_train['TotalBsmtSF'] > 0, 'HasBsmt'] = 1
```

```python
df_train.loc[df_train['HasBsmt'] == 1, 'TotalBsmtSF'] = np.log(df_train['TotalBsmtSF'])
```

```python
sns.distplot(df_train[df_train['TotalBsmtSF']>0]['TotalBsmtSF'], fit=norm)
fig = plt.figure()
res = stats.probplot(df_train[df_train['TotalBsmtSF']>0]['TotalBsmtSF'],plot=plt)
```
